```python
from sklearn.model_selection import train_test_split

random_md = train_md[train_md['primary_label'].isin(sample_bird_species)]
#display(random_md)
X, y = random_md.iloc[:,1:].values, random_md.iloc[:,:1].values
columns = train_md.columns

TEST_SPLIT = 0.2
RESERVED_SPLIT = 0.1

X_temp, X_test, y_temp, y_test = train_test_split(X, y, shuffle=True,test_size=TEST_SPLIT, random_state=None,stratify=y)
X_train, X_reserved, y_train, y_reserved = train_test_split(X_temp, y_temp, shuffle=True,test_size=RESERVED_SPLIT, random_state=None,stratify=y_temp)
X_train

tmp_df = pd.DataFrame(y_train, columns=columns[:1])
tmp_df2 = pd.DataFrame(X_train, columns=columns[1:])
train_df = pd.concat([tmp_df, tmp_df2], axis=1)

tmp_df = pd.DataFrame(y_test, columns=columns[:1])
tmp_df2 = pd.DataFrame(X_test, columns=columns[1:])
test_df = pd.concat([tmp_df, tmp_df2], axis=1)

tmp_df = pd.DataFrame(y_reserved, columns=columns[:1])
tmp_df2 = pd.DataFrame(X_reserved, columns=columns[1:])
reserved_df = pd.concat([tmp_df, tmp_df2], axis=1)
display(reserved_df)

# Write dataframes to files
train_df.to_csv('./random_train_metadata.csv', index=False)
test_df.to_csv('./random_test_metadata.csv', index=False)
reserved_df.to_csv('./random_reserved_metadata.csv', index=False)
```
