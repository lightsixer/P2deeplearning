```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

BASE_PATH = "/content/drive/MyDrive/P2_DeepLearning/birdclef-2023"

train_meta_path = os.path.join(BASE_PATH, "train_metadata.csv")
train_md = pd.read_csv(train_meta_path)
train_md.head()
```

```python
sample_bird_species = ['afbfly1', 'afdfly1', 'chespa1', 'chewea1', 'grywrw1', 'luebus1', 'nobfly1', 'rocmar2', 'somgre1', 'somtit4']
random_md = train_md[train_md['primary_label'].isin(sample_bird_species)]
filtered_bird_species = ['barswa', 'cohmar1', 'combuz1', 'comsan', 'eaywag1', 'eubeat1', 'litegr', 'thrnig1', 'wlwwar', 'woosan']
filtered_md = train_md.query(f"rating>=4")[train_md['primary_label'].isin(filtered_bird_species)]
```

```python
ax = random_md['primary_label'].value_counts().plot(kind='bar', figsize=(12,6), title="Count for each randomly selected species")
ax.set_xlabel("Class label")
ax.set_ylabel("Samples")
```

```python
ax = train_md['primary_label'].value_counts().plot(kind='bar', figsize=(16,10), title="Count for each data species")
ax.set_xlabel("Class label")
ax.set_ylabel("Samples")
```
